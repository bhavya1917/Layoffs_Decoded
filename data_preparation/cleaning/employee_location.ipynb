```python
import requests
def get_country(location):

    url = 'https://countriesnow.space/api/v0.1/countries/population/cities'
    location = location[0].lower()
    print(location)
    payload = {'city': location}

    response = requests.post(url, json=payload)
    print(response)
    if response.status_code == 200:
        data = response.json()['data']
        country = data['country']
        return country
    else:
        return f"Error: {response.status_code}"



    
```

```python
#only allows 1 request per sec, doesn't work even after sleep
from geopy.geocoders import Nominatim
import time
def get_location(raw_loc):
#     print("after cleaning: ", raw_loc)
    try:
        time.sleep(2)
        geolocator = Nominatim(user_agent="layoffs-decoded")
        location = geolocator.geocode(raw_loc, exactly_one=True,language="english", namedetails=True, addressdetails=True)
        country = location.raw['address']['country']
    except:
        country = "NA"
    print("output: ", country)
    print()
    print()
    return country
```

```python
def string_cleaning(raw_loc):
    if raw_loc == 'nan':
        raw_loc = ""
    raw_loc = raw_loc.replace("remote","").replace("hybrid","").replace("anywhere","")
    raw_loc = raw_loc.replace(".","")
    return raw_loc
```

```python
import requests

def get_country_from_address(address):
    base_url = "https://nominatim.openstreetmap.org/search"
    params = {
        "q": address,
        "format": "jsonv2"
    }
    response = requests.get(base_url, params=params).json()
    if response:
        return response[0]["display_name"].split(',')[-1]
    return None
```
