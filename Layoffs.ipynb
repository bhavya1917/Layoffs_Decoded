```python
#CODE TO RUN NEXT

# merge with AlphaVantage

def retrieve_alpha_data(symbol, apikey):
  user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'

  income_url = "https://www.alphavantage.co/query?function=INCOME_STATEMENT&symbol=" + symbol + "&apikey=" + apikey
  res = requests.get(url=income_url, headers={'User-Agent': user_agent})
  data = res.json()
  income_statement_df = pd.DataFrame(data['annualReports'])

  balance_sheet_url = "https://www.alphavantage.co/query?function=BALANCE_SHEET&symbol=" + symbol + "&apikey=" + apikey
  res = requests.get(url=balance_sheet_url, headers={'User-Agent': user_agent})
  data = res.json()
  balance_sheet_df = pd.DataFrame(data['annualReports'])

  cash_flow_url = "https://www.alphavantage.co/query?function=CASH_FLOW&symbol=" + symbol + "&apikey=" + apikey
  res = requests.get(url=balance_sheet_url, headers={'User-Agent': user_agent})
  data = res.json()
  cash_flow_df = pd.DataFrame(data['annualReports'])

  # CONSULT WITH THE TEAM FIRST 
  
  # monthly_stock_price_url = "https://www.alphavantage.co/query?function=TIME_SERIES_MONTHLY&symbol=" + symbol + "&apikey=" + apikey
  # res = requests.get(url=monthly_stock_price_url, headers={'User-Agent': user_agent})
  # data = res.json()
  # cash_flow_df = pd.DataFrame(data['annualReports'])

  # filter all 3 using common fiscalDateEnding to only use data from 2018
  income_statement_df = income_statement_df[int(income_statement_df['fiscalDateEnding'].split('-')[0]) >= 2018]
  balance_sheet_df = balance_sheet_df[int(balance_sheet_df['fiscalDateEnding'].split('-')[0]) >= 2018]
  cash_flow_df = cash_flow_df[int(cash_flow_df['fiscalDateEnding'].split('-')[0]) >= 2018]

  # merge all 3 using fiscalDateEnding
  merged_df1 = income_statement_df.merge(balance_sheet_df, on='fiscalDateEnding')
  final_merged_df = merged_df1.merge(cash_flow_df, on='fiscalDateEnding')

  return final_merged_df

```
