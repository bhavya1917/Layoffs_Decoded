```python
def merge_frames(frame1, frame2):
    return pd.merge(frame1, frame2, how='inner', on=['feature_names'])
    
```

```python
def intersection(pearson_df, spearman_df, ridge_df, randomforest_df, elasticnet_df, decisiontree_df, lasso_df, backward_elimination_df, forward_elimination_df, stepwise_elimination_df):
    pearson_df['feature_names'] = pearson_df['feature_names'].str.strip()
    spearman_df['feature_names'] = spearman_df['feature_names'].str.strip()
    ridge_df['feature_names'] = ridge_df['feature_names'].str.strip()
    randomforest_df['feature_names'] = randomforest_df['feature_names'].str.strip()
    elasticnet_df['feature_names'] = elasticnet_df['feature_names'].str.strip()
    decisiontree_df['feature_names'] = decisiontree_df['feature_names'].str.strip()
    lasso_df['feature_names'] = lasso_df['feature_names'].str.strip()
    backward_elimination_df['feature_names'] = backward_elimination_df['feature_names'].str.strip()
    forward_elimination_df['feature_names'] = forward_elimination_df['feature_names'].str.strip()
    stepwise_elimination_df['feature_names']= stepwise_elimination_df['feature_names'].str.strip()
    
    temp_df = pd.merge(pearson_df, spearman_df, how='inner', on=['feature_names'])
    temp1_df = pd.merge(ridge_df, randomforest_df, how='inner', on=['feature_names'])
    temp2_df = pd.merge(forward_elimination_df, stepwise_elimination_df, how='inner', on=['feature_names'])

    dfs = [elasticnet_df, lasso_df, decisiontree_df]
    dfs1 = [backward_elimination_df]

    for i in dfs:
        temp1_df = merge_frames(temp1_df, i)
    
    for i in dfs1:
        temp2_df = merge_frames(temp2_df, i)
        
    return temp_df, temp1_df, temp2_df

result_df, result_df1, result_df2 = intersection(precovid_pearson_df, precovid_spearman_df, precovid_ridge_df, precovid_randomforest_df, precovid_elasticnet_df, precovid_decisiontree_df, precovid_lasso_df, precovid_backward_elimination_df, precovid_forward_elimination_df, precovid_stepwise_elimination_df)
```
